```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import NearestCentroid
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
```

```python
df = pd.read_csv('/mnt/home/jbielecki1/pilk.csv')
ilosc_atrybutow = len(df.columns)    # liczba wszystkich atrybutów 
ilosc_obiektow = len(df) 
ilosc_brakow = df.isna().sum()
print("Liczba atrybutów: " + str(ilosc_atrybutow))
print("Liczba obiektów: " + str(ilosc_obiektow))
print("Liczba braków: " + str(ilosc_brakow))
df.info()
```

```python
#Usuwanie zbędnych danych
df = df.drop(['Photo', 'Flag', 'Club Logo', 'Loaned From', 'Real Face'], axis=1)
#Usuwanie wierszy z brakującym klubem
df = df = df[pd.notnull(df['Club'])]
#Wypelnianie wierszy z brakujacymi danymi
df['Club'].fillna(value="Unemployed", inplace=True)
df['Preferred Foot'].fillna(value="Both", inplace=True)

#Wypełnienie wartościami średnimi brakujących pozycji
#df.apply(lambda x: x.fillna(x.mean()),axis=0)
df.fillna(df.mean(), inplace=True)
ilosc_brakow = df.isna().sum()
print("Liczba braków: " + str(ilosc_brakow))
```

```python
#Zmienność atrybutów 
atr = 'Age'
plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(df.loc[:,atr], bins = 17)
plt.title("Age")

atr = 'Overall'
plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(df.loc[:,atr], bins = 17)
plt.title("Overall")

atr = 'Potential'
plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(df.loc[:,atr], bins = 17)
plt.title("Potential")

df['Release Clause'].plot.hist(stacked=True, bins=30)

atr = 'BallControl'
plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(df.loc[:,atr], bins = 17)
plt.title("Ball Control")

atr = 'GKHandling'
plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(df.loc[:,atr], bins = 17)
plt.title("GKHandling")
```

```python
#Wartośći najmniejsze i najwieksze w kolumnach age, overall, potentail   zakresy
Minimalny_wiek=df['Age'].min()
Maksymalny_wiek=df['Age'].max()
Minimalny_Overall=df['Overall'].min()
Maksymalny_Overall=df['Overall'].max()
Minimalny_Potential=df['Potential'].min()
Maksymalny_Potential=df['Potential'].max()
print('Minimalny wiek      to: ' + str(Minimalny_wiek)+" natomiast Maksymalny: "+str(Maksymalny_wiek))
print('Minimalny Overall   to: ' + str(Minimalny_Overall)+" natomiast Maksymalny: "+str(Maksymalny_Overall))
print('Minimalny Potencjal to: ' + str(Minimalny_Potential)+" natomiast Maksymalny: "+str(Maksymalny_Potential))
```

```python
#Potencjał age >28 oraz age <28
Wiekszyf=df.loc[df['Age']>=28]
Mniejszyf=df.loc[df['Age']<28]
atr = 'Potential'
plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(Wiekszyf.loc[:,atr], bins = 17)
plt.title("Potencjał piłkarzy powyżej 28 roku")
plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(Mniejszyf.loc[:,atr], bins = 17)
plt.title("Potencjał piłkarzy poniżej 28 roku")
```

```python
#Korelacja potenciału do Overalla
plt.figure(figsize=(12,12),dpi = 200)
corr = df.iloc[:,2:55].corr()
a=sns.heatmap(corr,annot = corr)


```

```python
silnie_rozwiniete = ((corr >= 0.6).sum().sum() - 9) // 2 
slabo_rozwiniete = ((0.6 > corr) & (corr >= 0.2)).sum().sum() // 2
brak_korelacji = ((0.2 > corr) & (corr > -0.2)).sum().sum() // 2
slabo = ((-0.2 >= corr) & (corr > -0.6)).sum().sum() // 2
silnie_ujemnie = (corr <= -0.6).sum().sum() // 2
print("Silnie skorelowane:         " + str(silnie_rozwiniete))
print("Slabo skorelowane:          " + str(slabo_rozwiniete))
print("Brak korelacji:             " + str(brak_korelacji))
print("Slabo ujemnie skorelowane:  " + str(slabo))
print("Silnie ujemnie skorelowane: " + str(silnie_ujemnie))
```
